```python
new_df1 = fitbit.copy(deep=True)
new_df1.drop(new_df1.loc[new_df1['Minutes Sedentary']< 100].index, inplace=True)
new_df1.drop(new_df1.loc[new_df1["Steps"]< 300].index, inplace=True)
new_df1
```

```python
new_df1.columns
```

```python
new_df1.rename(columns={"Minutes Lightly Active": "Minutes_Lightly_Active",
                       "Minutes Very Active": "Minutes_Very_Active", 
                       "Minutes Sedentary": "Minutes_Sedentary"},
              inplace=True)
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))

sns.scatterplot(x=new_df1.Steps, y=new_df1.Minutes_Very_Active,ax=axs[1, 0],
                legend="auto").annotate("In general, the more minutes very active the more steps",
                                                      xy=(18, 150), 
                                                      xycoords='data', arrowprops=dict(facecolor='red', shrink=0.05))
sns.scatterplot(x=new_df1.Steps, y=new_df1.Minutes_Sedentary, ax=axs[0, 0], legend="full").title.set_text("Minutes_Sedentary / steps")
sns.scatterplot(x=new_df1.Minutes_Very_Active, y=new_df1.Minutes_Sedentary , ax=axs[0, 1])
sns.scatterplot(x=new_df1.Steps, y=new_df1.Minutes_Lightly_Active , ax=axs[1, 1])
```
